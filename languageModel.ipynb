```python
import nltk
from nltk import *
```

```python
text="It is a long established fact that a reader will be distracted by the readable content of a page when looking at its layout. The point of using Lorem Ipsum is that it has a more-or-less normal distribution of letters, as opposed to using 'Content here, content here', making it look like readable English. Many desktop publishing packages and web page editors now use Lorem Ipsum as their default model text, and a search for 'lorem ipsum' will uncover many web sites still in their infancy. Various versions have evolved over the years, sometimes by accident, sometimes on purpose"
```

```python
corpus=word_tokenize(text)

lower_corpus=[]

for w in corpus:
    if w.isalpha()==True:
        lower_corpus.append(w.lower())

vocab=set(lower_corpus)
```

```python
bigram_count={}
trigram_count={}

for i in range(len(lower_corpus)-2):
    bigram=(lower_corpus[i],lower_corpus[i+1])
    trigram=(lower_corpus[i],lower_corpus[i+1],lower_corpus[i+2])

    if bigram in bigram_count.keys():
        bigram_count[bigram]+=1
    else:
        bigram_count[bigram]=1
    

    if trigram in trigram_count.keys():
        trigram_count[trigram]+=1
    else:
        trigram_count[trigram]=1

```

```python
def next_word(inptt,bigram_count,trigram_count,vocab):
    token=word_tokenize(inptt.lower())
    last_bigram=token[-2:]

    #probablity

    vocab_probab={}

    for vocab_word in vocab:
        testTrigram=(last_bigram[0],last_bigram[1],vocab_word)
        testBigram=(last_bigram[0],last_bigram[1])

        testTrigram_count=trigram_count.get(testTrigram,0)
        testBigram_count=bigram_count.get(testBigram,0)

        probab=testTrigram_count/testBigram_count
        vocab_probab[vocab_word]=probab

    suggestion=sorted(vocab_probab.items(),key=lambda x:x[1],reverse=True)[:3]

    return suggestion
```

```python
next_word('many web sites',bigram_count,trigram_count,vocab)
```
