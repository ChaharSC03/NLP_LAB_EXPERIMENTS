```python
import nltk
from nltk import *
```

```python
text="There are many variations of passages of Lorem Ipsum available, but the majority have suffered alteration in some form, by injected humour, or randomised words which don't look even slightly believable. If you are going to use a passage of Lorem Ipsum, you need to be sure there isn't anything embarrassing hidden in the middle of text. All the Lorem Ipsum generators on the Internet tend to repeat predefined chunks as necessary, making this the first true generator on the Internet. It uses a dictionary of over 200 Latin words, combined with a handful of model sentence structures, to generate Lorem Ipsum which looks reasonable. The generated Lorem Ipsum is therefore always free from repetition, injected humour, or non-characteristic words etc."
```

```python
corpus=word_tokenize(text)
```

```python
mainCorpus=[]

for w in corpus:
    if w.isalpha()==True:
        mainCorpus.append(w.lower())
mainCorpus
```

```python
vocab=set(mainCorpus)
vocab
```

```python
bigram_counts={}
trigram_counts={}

for i in range(len(mainCorpus)-2):
    bigram=(mainCorpus[i],mainCorpus[i+1])
    trigram=(mainCorpus[i],mainCorpus[i+1],mainCorpus[i+2])

    if bigram in bigram_counts.keys():
        bigram_counts[bigram]+=1
    else:
        bigram_counts[bigram]=1
    
    if trigram in trigram_counts.keys():
        trigram_counts[trigram]+=1
    else:
        trigram_counts[trigram]=1
```

```python
def nextWord(inptt, bigram_counts, trigram_counts, vocab):

    token=word_tokenize(inptt.lower())
    lastWord=token[-2:]

    vocab_probab={}

    for vw in vocab:

        testTrigram=(lastWord[0],lastWord[1],vw)
        testbigram=(lastWord[0],lastWord[1])

        testTrigram_count=trigram_counts.get(testTrigram,0)
        testbigram_count=bigram_counts.get(testbigram,0)

        probab=testTrigram_count/testbigram_count

        vocab_probab[vw]=probab
    
    suggestion=sorted(vocab_probab.items(),key=lambda x:x[1], reverse=True)[:3]
    return suggestion
```

```python
nextWord("the internet", bigram_counts, trigram_counts, vocab)
```
